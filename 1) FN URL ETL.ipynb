```python
import os, sys, re, time, gc, csv, warnings

warnings.filterwarnings("ignore")

Notebook_start = time.time()

read_loc = 'dataset/Raw/Read/'
store_loc = 'dataset/Raw/'
save_loc = f'{store_loc}LIWC/'

read_file = 'URL_True' # URL_True, URL_False, True, Fake


from LIWC_Methods import Custom_Methods 

import pandas as pd
import numpy as np

from urllib.request import urlsplit
```

```python
CM = Custom_Methods()
```

```python
print(f'{read_loc}{read_file}.csv')
df_pol = pd.read_csv(f'{read_loc}politifact_data.csv',error_bad_lines=False, quotechar='"', thousands=',' ,low_memory=False)
df_pol['source'] = ''
df_pol['title'] = ''
df_pol['body'] = ''
df_pol['class'] = [0 if c > 2 else c for c in df_pol['class']]
df_pol.drop(columns='score',inplace=True)
keep_list = ['id','news_url','source','title','body','class']
df_pol = df_pol[keep_list]
# df_pol.drop(columns)
print(len(df_pol[df_pol['class']==0]),len(df_pol[df_pol['class']==1]))
df_pol
```

```python
drop_idx = df_pol.index[df_pol.news_url == ''].tolist()
drop_idx
```

```python
print(f'{read_loc}{read_file}.csv')
df_pol_f = pd.read_csv(f'{read_loc}politifact_fake.csv',error_bad_lines=False, quotechar='"', thousands=',' ,low_memory=False)
df_pol_f.drop(columns=['title','tweet_ids'],inplace=True)
df_pol_f['source'] = ''
df_pol_f['title'] = ''
df_pol_f['body'] = ''
df_pol_f['class'] = '0'

df_pol_f
```

```python
print(f'{read_loc}{read_file}.csv')
df_pol_t = pd.read_csv(f'{read_loc}politifact_real.csv',error_bad_lines=False, quotechar='"', thousands=',' ,low_memory=False)
df_pol_t.drop(columns=['title','tweet_ids'],inplace=True)
df_pol_t['source'] = ''
df_pol_t['title'] = ''
df_pol_t['body'] = ''
df_pol_t['class'] = 0

df_pol_t
```

```python
print(f'{read_loc}{read_file}.csv')
df_kag = pd.read_csv(f'{read_loc}Kaggle_News.csv',error_bad_lines=False, quotechar='"', thousands=',' ,low_memory=False)
df_kag['class'] = [0 if c == 1 else 1 for c in df_kag['Label']]
df_kag['id'] = [f'Fake_kag_{idx}' if df_kag['class'][idx] == 1 else f'True_kag_{idx}' for idx in df_kag.index]
df_kag['source'] = ''
col_dict = {'URLs':'news_url','Headline':'title','Body':'body'}
keep_list = ['id','news_url','source','title','body','class']
df_kag.rename(columns=col_dict,inplace=True)
df_kag = df_kag[keep_list]
df_kag
```

```python
URL_True = df_pol[df_pol['class']==0]
print(len(URL_True))
URL_True = URL_True.append(df_pol_t,ignore_index=True)
URL_True.reset_index(drop=True,inplace=True)
URL_True['id'] = [f'True_pol_{idx}' for idx in URL_True.index]
print(len(URL_True))
# URL_True = URL_True.append(df_gos_t,ignore_index=True)
# print(len(URL_True))
URL_True = URL_True.append(df_kag[df_kag['class']==0])
print(len(URL_True))
# Remove dups
URL_True = URL_True[URL_True['news_url'].duplicated()==False]
print('After drop dups:',len(URL_True))
drop_idx = URL_True.index[URL_True.news_url == ''].tolist()
URL_True.drop(index=drop_idx,inplace=True)
print('len drop idx',len(drop_idx))
URL_True.reset_index(drop=True,inplace=True)
URL_True['source'] = CM.get_url_base(URL_True,'news_url','source')
print(len(URL_True))
URL_True
```

```python
URL_Fake = df_pol[df_pol['class']==1]
print(len(URL_Fake))
URL_Fake = URL_Fake.append(df_pol_f,ignore_index=True)
URL_Fake.reset_index(drop=True,inplace=True)
URL_Fake['id'] = [f'Fake_pol_{idx}' for idx in URL_Fake.index]
# print(len(URL_Fake))
# URL_Fake = URL_Fake.append(df_gos_f,ignore_index=True)
print(len(URL_Fake))
URL_Fake = URL_Fake.append(df_kag[df_kag['class']==1])
print(len(URL_Fake))
# Remove dups
URL_Fake = URL_Fake[URL_Fake['news_url'].duplicated()==False]
print('After drop dups:',len(URL_Fake))
drop_idx = URL_Fake.index[URL_Fake.news_url == ''].tolist()
print('len drop idx',len(drop_idx))
URL_Fake.drop(index=drop_idx,inplace=True)
URL_Fake.reset_index(drop=True,inplace=True)
URL_Fake['source'] = CM.get_url_base(URL_Fake,'news_url','source')
print(len(URL_Fake))
URL_Fake
```

```python
# Save URL files
for file in [[URL_Fake,'URL_Fake'],[URL_True,'URL_True']]:
    loc = f'{read_loc}{file[1]}.csv'
    print(loc)
    file[0].to_csv(loc,mode='w',header=True,index=False)
```
