```python

for section in ['Body']: # ,'Title']:
    if section == 'Body':
        source_file_1 = 'Base_Body_Out.csv'
        source_file_2 = 'URL_Body_Out.csv'
    else:
        source_file_1 = 'Base_Title_Out.csv'
        source_file_2 = 'URL_Title_Out.csv'
    
    # Load files
    loc = f'{file_loc}{source_file_1}'
    print(loc)
    df_1 = pd.read_csv(loc)
    loc = f'{file_loc}{source_file_2}'
    df_2 = pd.read_csv(loc)
    df = df_1.append(df_2)
    df['class'] = [1 if 'Fake' in val else 0 for val in df.Filename]
    df['id'] = [val[:len(val)-4] for val in df.Filename]
    drop_col = ['Segment']
    df.drop(columns=drop_col,inplace=True)
    df = df.sample(frac=1) 
    df.dropna(axis=0,inplace=True)
    df = df[df.WC >= 10]
    df.reset_index(drop=True,inplace=True)
    drop_col = ['Filename','WC','class','id']
    
    
    
    
    # Scale results of not
    drop_col = ['Filename','WC','class','id']
    for scaled in [False]:
        print(f'Section: {section}, Scaled: {scaled}')
        if scaled:
            df[list(set(df.columns).difference(drop_col))] = CM.normalize(df[list(set(df.columns).difference(drop_col))])
        df.dropna(axis=1,inplace=True)

    
        # Get sample same number of Fake and True rows
        y = df['class']
        fake_idx= np.where(y==1)
        rand_fake_idx = random.sample(list(fake_idx[0]),min(len(y[y==0]),len(y[y==1])))
        true_idx = np.where(y==0)
        rand_true_idx = random.sample(list(true_idx[0]),min(len(y[y==0]),len(y[y==1])))
        keep_idx = list(np.concatenate((rand_fake_idx,rand_true_idx)))
        X = df.copy()
        X.drop(columns=drop_col,inplace=True) 

        true_idx = np.where(y==0)
        X_sample = df.iloc[keep_idx].copy()
        ID_col = df.id
        X_sample.drop(columns=drop_col,inplace=True)
        y_sample = y.iloc[keep_idx]

        X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.3, stratify=y_sample, random_state=42)
        print(X_train.shape,X_test.shape)
        print(len(y_train),len(y_test))

        trees = 100
        gc.collect()
        print('rf_clf')
        rf_clf = RandomForestClassifier(n_estimators=trees)
        rf_clf.fit(X_train, y_train)
        y_pred = rf_clf.predict(X_test)
        print("Statistical tests...")
        print("***************")
        a_score = round(accuracy_score(y_pred, y_test),3)
        print(f'Accuracy score: {a_score}')
        print("Confusion Matrix: ")
        cm = confusion_matrix(y_pred, y_test)
        print(cm)

        
# save the model to disk
filename = f'{file_loc}Stemmed_{file_type}/LIWC_body_model.pkl'
print(filename)
pickle.dump(rf_clf, open(filename, 'wb'))
```
